```python
import csv
import random

def loadDataset(filename, split, trainingSet=[], testSet=[]):
    with open(filename, 'r') as csvfile:
        lines = csv.reader(csvfile)
        dataset = list(lines)
        header = dataset[0]
        dataset = dataset[1:]  # Assuming the first row is header

        for row in dataset:
            # Convert numeric values from strings to floats (assuming columns 0-3 are numeric)
            for i in range(len(row)):
                row[i] = str(row[i])

            if random.random() > split:
                trainingSet.append(row)
            else:
                testSet.append(row)
```

```python
import math
def euclideanDistance(inst1 , inst2 , length):
    distance = 0
    for x in range(length):
        distance += pow((inst1[x] - inst2[x]) , 2)
    return math.sqrt(distance)
    
```

```python
data1 = [ 2,4,3,6 ,'a' ,'p']
data2 = [4,4,4,'b']
distance=euclideanDistance(data1 , data2 , 3)
print('Distance ' , repr(distance))
```

```python
import operator
def getNegibor(trainingSet , testInstance , k):
    distance = []
    length = len(trainingSet) - 1
    for x in range(len(trainingSet)):
        dist = euclideanDistance(testInstance ,trainingSet[x] , length)
        distance.append(trainingSet[x])
    distance.sort(key = operator.itemgetter(1))
    neigbor = []
    for x in range(k):
        neigbor.append(distance[x][0])
    return neigbor
    
```

```python
trainingSet = [[2,4,2,3,'p'],[4,4,55,6]]
testInstance = [5,5,5]
k = 2
neigbor = getNegibor(trainingSet , testInstance , 2)
print(neigbor)
```
